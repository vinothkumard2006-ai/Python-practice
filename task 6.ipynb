```python
data = [75, 80, 85, 90, 95, 100, 105, 110]

# Mean and standard error
mean = np.mean(data)
sem = stats.sem(data)

# 95% Confidence Interval
confidence_interval = stats.t.interval(
    0.95,  # confidence level
    len(data) - 1,  # degrees of freedom
    loc=mean,  # sample mean
    scale=sem   # standard error
)

print("Mean:", mean)
print("95% Confidence Interval:", confidence_interval)
```

```python
 imoprt numpy as np
data2 = np.array([78, 82, 88, 84, 90, 85, 89, 87])
test_value = 85
t_stat, p_value = stats.ttest_1samp(data2, test_value)
print("\nQ4: One-sample t-test")
print("t-statistic:", t_stat, "p-value:", p_value)
if p_value < 0.05:
    print("Reject H0: Mean ≠ 85")
else:
    print("Fail to reject H0: No evidence mean ≠ 85")
```

# Question 5

```python
class_A = np.array([88, 90, 85, 86, 87, 91, 89])
class_B = np.array([78, 79, 75, 80, 76, 77, 74])
t_stat2, p_value2 = stats.ttest_ind(class_A, class_B, equal_var=False)  # Welch’s test
print("\nQ5: Two-sample t-test (independent)")
print("t-statistic:", t_stat2, "p-value:", p_value2)
if p_value2 < 0.05:
    print("Reject H0: Significant difference between means")
else:
    print("Fail to reject H0: No significant difference")
```

# Question 6

```python
n = 50  
x = 30 
p_hat = x / n  
p0 = 0.5 
se = np.sqrt(p0 * (1 - p0) / n)

z = (p_hat - p0) / se

p_value = 1 - stats.norm.cdf(z)

result = {
    'z_statistic': z,
    'p_value': p_value,
    'significance_level': 0.05,
    'reject_null': p_value < 0.05,
}
print(result)
```
