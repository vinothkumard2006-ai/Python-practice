```python
data={"Name":["Alice","Bob","Charlie","David"],"Age":[25,None,23,None],"City":["Pune","Delhi",None,"Mumbai"]}
df=pd.DataFrame(data)
print(df)
print("\nMissing values from column:")
print(df.isna().sum())

df["Age"]=df["Age"].fillna(df["Age"].mean())
df.dropna(thresh=len(df.columns)-1,inplace=True)
print("\nHandling missing values:")
print(df)
```

# Question 7

```python
import matplotlib.pyplot as plt
data={"Name":["Alice","Bob","Charlie","David","Eva"],"Department":["HR","IT","Finance","HR","IT"]}
df=pd.DataFrame(data)
dept_count=df["Department"].value_counts()
plt.bar(dept_count.index,dept_count.values)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Employees per Department")
plt.show()
```

# Question 8

```python
import pandas as pd
data={"Name":["A","B","C","D","E"],"Age":[25,None,23,None,30],"City":["Pune",None,"Mumbai","Pune",None]}
df=pd.DataFrame(data)
print(df)
print("\nmissing values column")
print(df.isna().sum())
df["Age"]=df["Age"].fillna(df["Age"].median())
df["City"]=df["City"].fillna(df["City"].mode()[0])
print("\nHandling missing values for age and city")
print(df)
```

```python
import seaborn as sns
data={"Name":["Alice","Bob","Charlie","David","Alice"],"Department":["HR","IT","Finance","HR","HR"],"Salary":[50000,60000,55000,52000,50000]}
df=pd.DataFrame(data)
df_cleaned=df.drop_duplicates()
print("\nRemove duplicates")
print(df)
df.rename(columns={"Department":"Dept","Salary":"Income"},inplace=True)
print("\nRename columns")
print(df)
sns.barplot(data=df,x="Dept",y="Income")
plt.title("Average Income by Department")
plt.ylabel("Average Income")
plt.xlabel("Department")
plt.tight_layout()
plt.show()
```
